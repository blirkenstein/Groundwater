```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_excel('gw-levels.xlsx')

# Ensure that 'TimeStamp' column is in datetime format
df['TimeStamp'] = pd.to_datetime(df['TimeStamp'])

# Set 'TimeStamp' column as the index
df.set_index('TimeStamp', inplace=True)

# Display the first few rows of the dataframe
print(df.head())

# Plot the water level time series data
plt.figure(figsize=(10, 6))
plt.plot(df['WaterLevel'])
plt.title('Ground Water Level Time Series')
plt.xlabel('Date')
plt.ylabel('Water Level')
plt.grid(True)
plt.show()

# Plot the temperature time series data
plt.figure(figsize=(10, 6))
plt.plot(df['Temperature'], color='red')
plt.title('Temperature Time Series')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()

# Basic Analysis
print("Descriptive data")
print(df.describe())

print("check for zero values")
print(df.isnull().sum())

```

```python
# Create a histogram
df['WaterLevel'].hist()
plt.show()
```

```python
import plotly.graph_objects as go

# Plot the water level time series data
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=df.index, y=df['WaterLevel'], mode='lines', name='WaterLevel'))
fig1.update_layout(title='Ground Water Level Time Series', xaxis_title='Date', yaxis_title='Water Level')
fig1.show()

# Plot the temperature time series data
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=df.index, y=df['Temperature'], mode='lines', name='Temperature', line=dict(color='red')))
fig2.update_layout(title='Temperature Time Series', xaxis_title='Date', yaxis_title='Temperature')
fig2.show()
```
